```python
gender_count = crime_data["Vict Sex"].value_counts()
plt.barh(gender_count.index, gender_count.values, color='skyblue', edgecolor='black')
plt.title("Distribution of Victim Genders")
plt.xlabel("Count")
plt.ylabel("Gender")
plt.title ("Victim Genders")
```

```python
plt.figure(figsize=(8, 6)) 
plt.hist(crime_data["Vict Age"], bins=20, range = (0,100), color='skyblue', edgecolor='black', alpha = 0.5)
mean_age = crime_data["Vict Age"].mean()
median_age = crime_data["Vict Age"].median()
plt.axvline(x=mean_age, color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(x=median_age, color='green', linestyle='dashed', linewidth=1, label='Median')
plt.text(mean_age, plt.ylim()[1] * 0.95, f'Mean: {mean_age:.1f}', color='red', fontsize=12, va='center')
plt.text(median_age, plt.ylim()[1] * 0.9, f'Median: {median_age:.1f}', color='green', fontsize=12, va='center')
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.7)
plt.title ("Victim Age Distribution")
plt.legend()
plt.plot()
```

```python
weapon_counts = crime_data["Weapon Desc"].value_counts()


threshold_percentage = 2.0  
total_count = len(crime_data)
count_threshold = total_count * (threshold_percentage / 100.0)
modified_weapon_counts = weapon_counts.copy()
small_categories = weapon_counts[weapon_counts < count_threshold]
modified_weapon_counts = modified_weapon_counts.drop(small_categories.index)
modified_weapon_counts['Others'] = small_categories.sum()

colors = plt.cm.Paired.colors

plt.figure(figsize=(5, 5))
plt.pie(
    modified_weapon_counts,
    labels=modified_weapon_counts.index,   # Use weapon names (including "Others") as labels
    autopct="%1.1f%%",                    # Display percentage labels inside sections
    startangle=90,                        # Rotate the chart so that the labels start from the top
    colors=colors,                        # Use custom colors
    wedgeprops={'edgecolor': 'grey'},     # Add grey edge color to pie chart sections
    textprops={'fontsize': 10},            # Font size for labels
)

# Add a title to the pie chart
plt.title("Distribution of Weapon Types in Crime Data", fontsize=12)

# Display the pie chart
plt.axis("equal")  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
```

```python
offenses = crime_data["Crm Cd Desc"].value_counts()

threshold_percentage = 2.0  
total_count = len(crime_data)
count_threshold = total_count * (threshold_percentage / 100.0)
small_categories = offenses[offenses < count_threshold]
other_category_count = small_categories.sum()
modified_offense_counts = offenses.drop(small_categories.index)
modified_offense_counts['Others'] = other_category_count

colors = plt.cm.Paired.colors
plt.figure(figsize=(5, 5))
plt.pie(
    modified_offense_counts,             # Data to be plotted (modified offense counts)
    labels=modified_offense_counts.index, # Use modified offense names as labels
    autopct="%1.1f%%",                   # Display percentage labels inside sections
    startangle=90,                       # Rotate the chart so that the labels start from the top
    colors=colors,                       # Use custom colors
    wedgeprops={'edgecolor': 'grey'},    # Add grey edge color to pie chart sections
    textprops={'fontsize': 10},           # Font size for labels
)
plt.title("Distribution of Offenses Committed", fontsize = 30)
plt.axis("equal")
plt.show()
```

```python
crime_count_by_date = crime_data.groupby(['Crm Cd Desc', 'DATE OCC']).size().reset_index(name='Count')
crime_count_by_date['DATE OCC'] = pd.to_datetime(crime_count_by_date['DATE OCC'])

crime_count_by_date = crime_count_by_date.sort_values('DATE OCC')
plt.figure(figsize=(10, 6))

unique_crime_types = crime_count_by_date['Crm Cd Desc'].unique()

for crime_type in unique_crime_types:
    crime_type_data = crime_count_by_date[crime_count_by_date['Crm Cd Desc'] == crime_type]
    plt.plot(crime_type_data['DATE OCC'], crime_type_data['Count'], linestyle='-', marker='o', markersize=4, label=crime_type)


plt.title('Crime Occurrence Over Time by Offense Type')
plt.xlabel('Date (year/month)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
crime_count_by_date = crime_data.groupby(['Crm Cd Desc', 'DATE OCC']).size().reset_index(name='Count')
crime_count_by_date['DATE OCC'] = pd.to_datetime(crime_count_by_date['DATE OCC'])
false_imprisonment_data = crime_count_by_date[crime_count_by_date['Crm Cd Desc'] == "FALSE IMPRISONMENT"]
false_imprisonment_data = false_imprisonment_data.sort_values('DATE OCC')

plt.figure(figsize=(10, 6))
plt.plot(false_imprisonment_data['DATE OCC'], false_imprisonment_data['Count'], linestyle='-', marker='o', markersize=4)
plt.title('Crime Occurrence Over Time for "False Imprisonment"')
plt.xlabel('Date (year/month)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
crime_count_by_date = crime_data.groupby('DATE OCC').size().reset_index(name='Count')

crime_count_by_date['DATE OCC'] = pd.to_datetime(crime_count_by_date['DATE OCC'])
crime_count_by_date = crime_count_by_date.sort_values('DATE OCC')

plt.figure(figsize=(10, 4))
plt.plot(crime_count_by_date['DATE OCC'], crime_count_by_date['Count'], linestyle='-', marker='o', markersize=4, label="Crime Count")
plt.title('Crime Occurrence Over Time')
plt.xlabel('Date (year/month)')
plt.ylabel('Count')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
premis_counts = crime_data["Premis Desc"].value_counts()
plt.figure(figsize=(6, 6))

threshold_percentage = 5
total_count = premis_counts.sum()
small_values = premis_counts[premis_counts / total_count * 100 < threshold_percentage]
premis_counts = premis_counts[~premis_counts.index.isin(small_values.index)]
premis_counts["Others"] = small_values.sum()

premis_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title("Distribution of Premises Descriptions")
plt.legend(labels=premis_counts.index, loc="upper right")
plt.axis('equal')
plt.show()
```

```python
geometry = gp.points_from_xy(crime_data['LON'], crime_data['LAT'])
gdf_points = gp.GeoDataFrame(crime_data, geometry=geometry)
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xlim(gdf.total_bounds[0], gdf.total_bounds[2])
ax.set_ylim(gdf.total_bounds[1], gdf.total_bounds[3])
gdf.plot(ax=ax)
gdf_points.plot(ax=ax, color='red', markersize=1)

labeled_suburbs = {}

for idx, row in gdf.iterrows():
    suburb_name = row['NAME']  
    if suburb_name not in labeled_suburbs:
        suburb_centroid = row['geometry'].centroid
        ax.annotate(suburb_name, 
                    xy=(suburb_centroid.x, suburb_centroid.y),
                    xytext=(3, 3),  # Offset for the label
                    textcoords="offset points",
                    fontsize=5,
                    color='blue',  # Color for the label
                    horizontalalignment='left')
        labeled_suburbs[suburb_name] = True
plt.title('Los Angeles City Annexations with Highlighted Crime Data Points and Suburb Names')
plt.show()
```
